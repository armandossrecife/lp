### Definindo as coordenadas de pontos de um gráfico

x >= 1 e x <= 4, onde x ∈ Inteiros

y >= 1 e x <= 4, onde y ∈ Inteiros

```python
# Podemos definir um conjunto de valores (x, y) para plotar um gráfico
x = [1, 2, 3, 4] 
y = [1, 4, 2, 3]
```

```python
print(f'x: {x}')
print(f'y: {y}')
```

```python
# Exemplo de funcao que retorna uma lista dos pontos que serão plotados no gráfico
def f(x,y):
  coordenadas_xy = []
  tamanho = len(x)
  for i in range(tamanho):
    coordenadas_xy.append((x[i], y[i]))
  return coordenadas_xy
```

```python
minhas_coordenadas = f(x,y)
print(f'Lista de coordenadas {minhas_coordenadas}')
print(f'Qtd de coordanadas {len(minhas_coordenadas)}')
```

Obs: por questão de performance o Matplotlib trabalha com vetores numéricos via biblioteca [Numpy](https://numpy.org/).

## Exemplos usando dados gerados pelo automaticamente pelo Numpy

Função numpy.linspace

Retorna números uniformemente espaçados em um intervalo especificado.

https://numpy.org/doc/stable/reference/generated/numpy.linspace.html#numpy-linspace

```python
# Gera um vetor com 100 valores entre 0 e 2
x = np.linspace(0, 2, 100) 
print(x)
print(len(x))
```

```python
# Gera uma figura (fig) e os eixos (ax)
fig, ax = plt.subplots()

ax.plot(x, x, label='Função linear')  # Plot some data on the axes. f(x)
ax.plot(x, x**2, label='Função quadrática')  # Plot more data on the axes... f(xˆ2)
ax.plot(x, x**3, label='Função cúbica')  # ... and some more. f(xˆ3)

ax.set_xlabel('eixo x')  # Add an x-label to the axes.
ax.set_ylabel('eixo y')  # Add a y-label to the axes.
ax.set_title("Gráfico de funções")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

```python
# y1 = f(x)
# y2 = f(xˆ2)
# y3 = f(xˆ3)

y1 = x
print(f'f(x): {y1}')
print(f'Qtd de y1: {len(y1)}')
print('')

y2 = x*x
print(f'f(xˆ2): {y2}')
print(f'Qtd de y2: {len(y2)}')
print('')

y3 = x*x*x
print(f'f(xˆ3): {y3}')
print(f'Qtd de y3: {len(y3)}')
```

Destacando um ponto no gráfico

função ax.annotate(..) 

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.annotate.html

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t) # Função periódica 2pi*t 
line, = ax.plot(t, s, lw=2)

# Anotation
ax.annotate('local max (2,1)', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2);
```

### Gráfico de pizza

categorias (itens da pizza)

valores (devem somar 100 para representar 100% da pizza)

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
produtos = 'Laranja', 'Limão', 'Banana', 'Maçã'
quantidades = [20, 30, 40, 10]

print(f'Produtos: {produtos}')
print(f'Quantidades: {quantidades}')
```

```python
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Limão')

fig1, ax1 = plt.subplots()

ax1.pie(quantidades, explode=explode, labels=produtos, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Gráfico em 3d

Exemplo básico de um cubo

```python
from mpl_toolkits.mplot3d import Axes3D

# mpl_toolkits: fornece algumas ferramentas básicas de plotagem 3D (dispersão, surf, linha, malha). 
# É uma coleção de classes auxiliares para exibir eixos 3D no Matplotlib.
```

```python
# Nesta etapa, estamos selecionando o eixo 3D da dimensão X =3, Y=3, Z=3
axes = [3, 3, 3]
 
# np.ones() passa as dimensões do cubo.
# Cria os dados para serem plotados
data = np.ones(axes, dtype=np.bool)
 
# Controle de transparência
# Selecionamentos a opacidade da cor como alfa = 0,65 (varia de 0,0 a 1,0)
alpha = 0.65
 
# Controle de cores
# Passamos a dimensão dos eixos (ou seja, 3, 3, 3) + número de faces 
# para o cubo (ou seja, 0-4) na função np.empty() depois disso, passamos a combinação de cores 
# e opacidade para cada face do cubo .
colors = np.empty(axes + [4], dtype=np.float32)
 
colors[:] = [1, 0, 0, alpha]  # red
# colors[1] = [0, 1, 0, alpha] # green
# colors[2] = [0, 0, 1, alpha] # blue
# colors[3] = [1, 1, 0, alpha] # yellow
# colors[4] = [1, 1, 1, alpha] # grey

# Usamos a função figure() da biblioteca matplotlib que é usada para criar uma nova figura. 
# Depois disso, usamos o método add_subplot() para adicionar um eixo à figura como 
# tridimensional (ou seja, projeção = '3d' ) parte de um arranjo de subtrama. Tem 3 argumentos.
## O número de linhas na grade,
## O número de colunas na grade e,
## A posição na qual a nova subparcela deve ser colocada.

# Plota a figura (fig)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
# Voxels é usado para customizações do
# tamanhos, posições e cores.
ax.voxels(data, facecolors=colors)
```

### Esquemas de interpolação de matriz

A interpolação calcula qual a cor ou valor de um pixel "deveria" ser, de acordo com diferentes esquemas matemáticos. (Redimensionamento de imagem, criação de filtros em imagens, e outras transformações).

Um lugar comum em que isso acontece é quando você redimensiona uma imagem. O número de pixels muda, mas você deseja as mesmas informações. Como os pixels são discretos, falta espaço. A interpolação é como você preenche esse espaço. É por isso que suas imagens às vezes parecem pixelizadas quando você as amplia. O efeito é mais pronunciado quando a diferença entre a imagem original e a imagem expandida é maior. 

Vamos pegar nossa imagem e reduzi-la. Estamos efetivamente descartando pixels, mantendo apenas alguns selecionados. Agora, quando plotamos, esses dados são ampliados para o tamanho da tela. Os pixels antigos não estão mais lá e o computador precisa desenhar pixels para preencher esse espaço.

Usaremos a biblioteca Pillow que usamos para carregar a imagem também para redimensionar a imagem.

```python
from PIL import Image

img = Image.open(path_image)
img.thumbnail((64, 64))  # resizes image in-place
imgplot = plt.imshow(img)
```
